```python
df = df.dropna(subset=['Car Name'])
```

```python
df = df.dropna(subset=['Year'])
df['Year'] = df['Year'].astype(int)
```

```python
df['Location'] = df['Location'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Unnamed: 0'], errors='ignore')
```

```python
df
```

```python
df = df.drop_duplicates()
```

```python
df.dtypes
```

```python
cat_cols = ['Owner','Fuel', 'Location', 'Drive', 'Type']
df[cat_cols] = df[cat_cols].astype('category')
```

```python
df.dtypes
```

```python
CURRENT_YEAR = 2025
df['car_age'] = CURRENT_YEAR - df['Year']
```

```python
df['age_bucket'] = pd.cut(
    df['car_age'],
    bins=[0, 3, 6, 10, 20],
    labels=['0-3 yrs', '4-6 yrs', '7-10 yrs', '10+ yrs']
)
```

```python
df['price_per_km'] = df['Price'] / df['Distance']
```

```python
df
```

# Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

### Price Distribution

```python
sns.histplot(df['Price'], bins=50, kde=True)
plt.title("Distribution of Used Car Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

### Price vs Car Age

```python
sns.boxplot(x='age_bucket', y='Price', data=df)
plt.title("Price Distribution by Car Age")
plt.xlabel("Car Age")
plt.ylabel("Price")
plt.show()
```

### Price vs KM Driven

```python
sns.regplot(x='Distance', y='Price', data=df, scatter=False, color='red')
plt.title("Price vs KM Driven (Trend)")
plt.show()
```

### Fuel Type vs Price

```python
sns.boxplot(x='Fuel', y='Price', data=df)
plt.title("Price Distribution by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Price")
plt.show()
```

### Owner Count vs Price

```python
sns.boxplot(x='Owner', y='Price', data=df)
plt.title("Price vs Number of Owners")
plt.xlabel("Owner Count")
plt.ylabel("Price")
plt.show()
```

### Car Type vs Price

```python
sns.boxplot(x='Type', y='Price', data=df)
plt.title("Price Distribution by Car Type")
plt.xlabel("Car Type")
plt.ylabel("Price")
plt.show()
```

### Price per KM

```python
sns.boxplot(x='Fuel', y='price_per_km', data=df)
plt.ylim(0, 100)
plt.title("Price per KM by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Price per KM")
plt.show()
```
